```python
import model_loader
import pipeline
from PIL import Image
from pathlib import Path
from transformers import CLIPTokenizer
import torch
import os

ALLOW_CUDA = False
ALLOW_MPS  = False

if torch.cuda.is_available() and ALLOW_CUDA:
    DEVICE = "cuda"
elif (torch.backends.mps.is_available()) and ALLOW_MPS:
    DEVICE = "mps"
else:
    DEVICE = "cpu"

print(f"Using device: {DEVICE}")

BASE_DIR = r"C:\Users\Pavan Kumar\Desktop\umar_jamil\diffusion_scratch"

tokenizer = CLIPTokenizer(
    os.path.join(BASE_DIR, "data", "vocab.json"),
    merges_file=os.path.join(BASE_DIR, "data", "merges.txt")
)
model_file = os.path.join(BASE_DIR, "data", "v1-5-pruned-emaonly.ckpt")
models = model_loader.preload_models_from_standard_weights(model_file, DEVICE)

prompt = (
    "A highly detailed, ultra-realistic portrait of a young woman standing in soft "
    "golden sunlight, cinematic lighting, 8k resolution, sharp focus, intricate textures, "
    "natural skin tones, shallow depth of field, professional photography, volumetric light, "
    "beautiful atmosphere, bokeh background, award-winning photo"
)
uncond_prompt = ""
do_cfg = True
cfg_scale = 8

input_image = None
strength = 0.9

sampler = "ddpm"
num_inference_steps = 20 if DEVICE == "cpu" else 50
seed = 42

output_image = pipeline.generate(
    prompt=prompt,
    uncond_prompt=uncond_prompt,
    input_image=input_image,
    strength=strength,
    do_cfg=do_cfg,
    cfg_scale=cfg_scale,
    sampler_name=sampler,
    n_inference_steps=num_inference_steps,
    seed=seed,
    models=models,
    device=DEVICE,
    idle_device="cpu",
    tokenizer=tokenizer,
)

output_path = Path("output.png")
Image.fromarray(output_image).save(output_path)
print(f"Image saved to {output_path}")
Image.fromarray(output_image).show()
```
